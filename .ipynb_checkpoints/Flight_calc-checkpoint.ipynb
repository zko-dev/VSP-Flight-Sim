# VSP Flight Sim - Trim state convergence study

Readme: 
This code is used in pair with OPENVSP to generate a rough estimate to the performance of Group 1 Fixed Wing UAV under steady state scenarios. 

# Steps
1. If initial run, do setup.sh to install/update requirements.txt
2. From CAD model, determine MTOW, CG into this Flight_Calc.ipynb
    - Modify existing .vspaero file to setup for cfd
    - 
3. run_sim.sh
    - launches openvsp GUI
    - Generate degenGeom from .vsp3 using launch_openvsp.sh
    - runs batch sweep and triggers vspaero solver
    - outputs vsp_aero_result.csv in /output
4. Flight_Calc.ipynb extracts vsp_aero_result.csv into custom result display
    - Runs propulsion trade study

# Setup

```python
# === Full Code with Torque Penalty Based on Propeller Pitch ===
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import brentq

# === Functions ===
def air_density_at_altitude(h):
    T0 = 288.15
    P0 = 101325
    L = 0.0065
    R = 287.058
    g = 9.80665
    T = T0 - L * h
    P = P0 * (1 - L * h / T0) ** (g / (R * L))
    return P / (R * T)

# === Conditions ===
altitude = 16 #Davis altitude
rho = air_density_at_altitude(altitude)

# === Weight Estimation ===
parallel_batt_count = 2 #4s2p
serial_batt_count = 4
W_batt = 72*parallel_batt_count*serial_batt_count #4s2p
W_fuselage = 980
n_forward_rotors = 4
W_motor = n_forward_rotors * 50
W_MTOW = W_batt+W_fuselage + W_motor
W_batt_empty_frac = W_batt/W_MTOW

# === Aircraft Spec ===
W = W_MTOW / 1000 * 9.81 #N conversion
b = 0.15
Span = 1.0
Sref = b * Span
WS_ratio = W/Sref

# === Battery & Propulsion ===
battery_voltage = 4.0*serial_batt_count
battery_capacity_Ah = 4.2*parallel_batt_count  #Molicel P42A
battery_discharge_rate = 45*parallel_batt_count
usable_energy_fraction = 0.80
battery_energy_J = battery_capacity_Ah * battery_voltage * 3600
usable_energy_J = battery_energy_J * usable_energy_fraction
usable_energy_Wh = usable_energy_J / 3600
aux_current = 1.1
aux_power = aux_current * battery_voltage
prop_efficiency = 0.80

# === Powertrain Selection ===
KV = 1500 #https://emaxmodel.com/products/emax-eco-ii-series-2807-1300kv-1700kv-1500kv-brushless-motor-for-rc-drone-fpv-racing
pitch_in = 5.7
D_m = 7 * 0.0254
CT = 0.10
RPM_max = KV * battery_voltage 
n = RPM_max / 60             
V_pitch = pitch_in * 0.0254 * n
V_tip = D_m/2*n*(2*np.pi)

# === Aerodynamics ===
CD0 = 0.0328 #clean airfoil: 0.02812
CL0 = 0.957353146185
CL_max = 2.2 #s1223 low Re Airfoil
CL_alpha_per_deg = (1.062878605311-1.883487599524)/10  # Lift slope (per degree) from VSPAERO 0 deg to 10 deg
AR = Span**2 / Sref
V = np.linspace(4, V_pitch * 0.999, 600)
CL = (2 * W) / (rho * V**2 * Sref)
alpha_deg = (CL - CL0) / CL_alpha_per_deg
e_baseline = 1.78 * (1 - 0.045 * AR**0.68) - 0.64
e_wingtip_attenuation = e_baseline * 1.0 #1.15  Example 15% improvement sinngie et al
k = 1 / (np.pi * e_wingtip_attenuation * AR)
CD = CD0 + k * CL**2
LD_ratio = CL / CD
T_required = 0.5 * rho * V**2 * Sref * CD

```

# Result

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import plotly.graph_objects as go

# === Thrust using disk advancing ratio model ===
def estimate_thrust(V, RPM, D, CT, pitch):
    n = RPM / 60
    V_pitch = pitch * 0.0254 * n
    T_static = CT * rho * n**2 * D**4
    T = T_static * (1 - V / V_pitch)**2
    T[V >= V_pitch] = 0
    return T
T_avail = estimate_thrust(V, RPM_max, D_m, CT, pitch_in) *n_forward_rotors
def thrust_diff(v):
    cl = (2 * W) / (rho * v**2 * Sref)
    cd = CD0 + k * cl**2
    t_req = 0.5 * rho * v**2 * Sref * cd
    V_pitch_local = pitch_in * 0.0254 * (RPM_max / 60)
    T_static = CT * rho * (RPM_max / 60)**2 * D_m**4
    T_av = n_forward_rotors*T_static * (1 - v / V_pitch_local)**2 if v < V_pitch_local else 0
    return T_av - t_req
V_max = brentq(thrust_diff, pitch_in, V_pitch)
V_stall = np.sqrt((2 * W) / (rho * Sref * CL_max))

# === Torque Correction ===
Cq0 = 0.045 #Check these assumptions
p0 = 4.5
alpha_cq = 0.08
pitch_correction = 1 + alpha_cq * (pitch_in - p0)
thrust_ratio = np.divide(T_required, T_avail, out=np.zeros_like(T_required), where=T_avail > 0)
thrust_ratio = np.clip(thrust_ratio, 0, 1)
I_thrust = thrust_ratio * battery_discharge_rate
I_corrected = I_thrust * pitch_correction + aux_current * n_forward_rotors
P_actual = I_corrected * battery_voltage + aux_power
endurance_sec = usable_energy_J / P_actual
endurance_min = endurance_sec / 60
range_km = (V * endurance_sec) / 1000
idx_best = np.argmax(LD_ratio)
V_aero = V[idx_best]
I_best = I_corrected[idx_best]
range_best = range_km[idx_best]
endurance_best = endurance_min[idx_best]
I_stall = np.interp(V_stall, V, I_corrected)
I_max = np.interp(V_max, V, I_corrected)

# === Endurance vs Current Curve ===
I_vals = np.linspace(1, battery_discharge_rate, 500)
P_vals = I_vals * battery_voltage + aux_power
endurance_vals_min = usable_energy_Wh / (P_vals / 60)
endurance_best_lookup = np.interp(I_best, I_vals, endurance_vals_min)

# === Plot: Lift-to-Drag Ratio vs Velocity with AoA Annotations ===
plt.figure(figsize=(10, 6))
plt.plot(V, LD_ratio, color='teal', label='L/D = CL/CD')
plt.axvline(V_stall, color='red', linestyle='--', label=f"V_stall = {V_stall:.2f} m/s ({V_stall*3.6:.1f} km/h)")
plt.axvline(V_aero, color='green', linestyle='--', label=f"V_aero (L/D max) = {V_aero:.2f} m/s ({V_aero*3.6:.1f} km/h)")
plt.axvline(V_max, color='black', linestyle='--', label=f"V_max = {V_max:.2f} m/s ({V_max*3.6:.1f} km/h)")
max_LD = np.max(LD_ratio)
plt.plot(V_aero, max_LD, 'o', color='blue', markersize=7)
plt.text(V_aero, max_LD, f"Max L/D = {max_LD:.2f}", verticalalignment='bottom', color='blue')
# Annotate AoA every 20 points
for i in range(0, len(V), 20):
    plt.annotate(f"{-alpha_deg[i]:.1f}°", (V[i], LD_ratio[i]),
                 textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8, color='gray')
plt.xlabel("Airspeed (m/s)")
plt.ylabel("L/D Ratio")
plt.title("Lift-to-Drag Ratio vs Airspeed (with AoA Labels)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# === Plot: Thrust Required vs Available ===
plt.figure(figsize=(10, 6))
plt.plot(V, T_required, label='Thrust Required', color='blue')
plt.plot(V, T_avail, label='Thrust Available', color='orange')
plt.axvline(V_stall, linestyle='--', color='red', label=f"V_stall = {V_stall:.2f} m/s ({V_stall*3.6:.1f} km/h)")
plt.axvline(V_aero, linestyle='--', color='green', label=f"V_aero = {V_aero:.2f} m/s ({V_aero*3.6:.1f} km/h)")
plt.axvline(V_max, linestyle='--', color='black', label=f"V_max = {V_max:.2f} m/s ({V_max*3.6:.1f} km/h)")
T_best = np.interp(V_aero, V, T_required)
plt.plot(V_aero, T_best, 'o', color='green')
plt.text(V_aero, T_best, f"{T_best:.2f} N", color='green', verticalalignment='bottom')
plt.xlabel("Airspeed (m/s)")
plt.ylabel("Thrust (N)")
plt.title("Thrust Required vs Available")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# === Plot: Corrected Current vs Airspeed ===
plt.figure(figsize=(10, 6))
plt.plot(V, I_corrected, color='darkorange', label='Corrected Current')
plt.axvline(V_stall, linestyle='--', color='red', label=f"V_stall = {V_stall:.2f} m/s ({V_stall*3.6:.1f} km/h)")
plt.plot(V_stall, I_stall, 'o', color='red')
plt.text(V_stall, I_stall, f"{I_stall:.2f} A", color='red', verticalalignment='bottom')
plt.axvline(V_aero, linestyle='--', color='green', label=f"V_aero = {V_aero:.2f} m/s ({V_aero*3.6:.1f} km/h)")
plt.plot(V_aero, I_best, 'o', color='green')
plt.text(V_aero, I_best, f"{I_best:.2f} A", color='green', verticalalignment='bottom')
plt.axvline(V_max, linestyle='--', color='black', label=f"V_max = {V_max:.2f} m/s ({V_max*3.6:.1f} km/h)")
plt.plot(V_max, I_max, 'o', color='black')
plt.text(V_max, I_max, f"{I_max:.2f} A", color='black', verticalalignment='bottom')
plt.xlabel("Airspeed (m/s)")
plt.ylabel("Current Draw (A)")
plt.title("Corrected Current vs Airspeed")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# === Endurance Plot with Key Points ===
plt.figure(figsize=(10, 6))
plt.plot(I_vals, endurance_vals_min, color='darkblue', label='Endurance vs Current')
points = {
    "Stall": I_stall,
    "V_aero": I_best,
    "25% thrust": 0.25 * battery_discharge_rate,
    "75% thrust": 0.75 * battery_discharge_rate,
    "Max (100%)": battery_discharge_rate
}
for label, current in points.items():
    endurance = np.interp(current, I_vals, endurance_vals_min)
    plt.axvline(current, linestyle='--', label=f"{label} = {current:.1f} A", alpha=0.6)
    plt.plot(current, endurance, 'o', label=f'{label}: {endurance:.1f} min')
plt.xlabel("Current (A)")
plt.ylabel("Endurance (min)")
plt.title("Endurance vs Current (with Key Points)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# === Plot: Range vs Velocity (Corrected) with Endurance Annotations ===
plt.figure(figsize=(10, 6))
plt.plot(V, range_km, color='darkgreen', label='Range Curve')
endurance_at_stall = np.interp(V_stall, V, endurance_min)
endurance_at_best = endurance_best
endurance_at_max = np.interp(V_max, V, endurance_min)
plt.axvline(V_stall, color='red', linestyle='--', label=f"V_stall = {V_stall:.2f} m/s ({V_stall*3.6:.1f} km/h)")
plt.plot(V_stall, np.interp(V_stall, V, range_km), 'o', color='red')
plt.text(V_stall, np.interp(V_stall, V, range_km),
         f"{np.interp(V_stall, V, range_km):.2f} km\n{endurance_at_stall:.1f} min",
         color='red', verticalalignment='bottom')
plt.axvline(V_aero, color='green', linestyle='--', label=f"V_aero = {V_aero:.2f} m/s ({V_aero*3.6:.1f} km/h)")
plt.plot(V_aero, range_best, 'o', color='green')
plt.text(V_aero, range_best,
         f"{range_best:.2f} km\n{endurance_at_best:.1f} min",
         color='green', verticalalignment='bottom')
plt.axvline(V_max, color='black', linestyle='--', label=f"V_max = {V_max:.2f} m/s ({V_max*3.6:.1f} km/h)")
plt.plot(V_max, np.interp(V_max, V, range_km), 'o', color='black')
plt.text(V_max, np.interp(V_max, V, range_km),
         f"{np.interp(V_max, V, range_km):.2f} km\n{endurance_at_max:.1f} min",
         color='black', verticalalignment='bottom')
idx_max_range = np.argmax(range_km)
V_range_max = V[idx_max_range]
range_max = range_km[idx_max_range]
endurance_range_max = endurance_min[idx_max_range]
plt.plot(V_range_max, range_max, 'D', color = 'blue', label=f"V_range = {V_range_max:.2f} m/s ({V_range_max*3.6:.1f} km/h)")
plt.text(V_range_max, range_max, f"{range_max:.2f} km\n{endurance_range_max:.1f}min", color = 'blue', verticalalignment = 'bottom')
plt.xlabel("Airspeed (m/s)")
plt.ylabel("Range (km)")
plt.title("Range vs Airspeed with at Trimmed Flight")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# === Static Thrust Model & Hover Analysis with Pitch Correction ===
eta_prop = 0.7     # Propeller efficiency
eta_motor = 0.85   # Motor efficiency
I_0 = 0.8          # No-load current
alpha_hover = 0.10
p0 = 4.5           # Reference pitch for correction

# === Model: Thrust from Current (reversed momentum theory) ===
def thrust_from_current(I_motor, D_m, V_batt, rho, pitch_in, p0, alpha_hover,
                        eta_prop=0.7, eta_motor=0.85, I_0=0.8):
    I_base = I_motor / (1 + alpha_hover * (pitch_in - p0))
    P_elec = (I_base - I_0) * V_batt
    A = np.pi * (D_m**2) / 4
    P_ideal = P_elec * eta_prop * eta_motor
    T = (P_ideal * np.sqrt(2 * rho * A))**(2/3)
    return T
I_empirical = np.array([2.6, 5, 7.9, 11.5, 15.9, 53.3])  # A
I_empirical = I_empirical * 14.8/21 # Scaling to 4s
T_empirical = np.array([300, 500, 700, 900, 1100, 2420])  # grams (already scaled)
T_empirical = T_empirical * 14.8/21 # Scaling to 4s
I_test = np.linspace(1, I_empirical[-1], 300)
T_pred = thrust_from_current(I_test, D_m, battery_voltage, rho, pitch_in, p0, alpha_hover,
                             eta_prop=eta_prop, eta_motor=eta_motor, I_0=I_0)
T_pred_grams = T_pred * 1000 / 9.81  # N to g
plt.figure(figsize=(10, 6))
plt.plot(I_test, T_pred_grams, label='Model Prediction', color='green')
plt.scatter(I_empirical, T_empirical, color='red', label='Emax eco2807 1500kv | 7x3.5x3', zorder=5)
plt.xlabel("Current (A)")
plt.ylabel("Static Thrust (g)")
plt.title("Static Thrust vs Current (Single Rotor, scaled to 14.8V)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# === Hovering Subsection (Momentum + Pitch Correction) ===
def hover_current_momentum(thrust_N, D_m, V_batt, rho, pitch_in, p0, alpha_hover,
                           eta_prop, eta_motor, I_0=0.8):
    A = np.pi * (D_m**2) / 4
    P_ideal = thrust_N**1.5 / np.sqrt(2 * rho * A)        # Ideal induced power
    P_elec = P_ideal / (eta_prop * eta_motor)             # Electrical power input
    I_base = P_elec / V_batt + I_0                        # Base current estimate
    pitch_correction = 1 + alpha_hover * (pitch_in - p0)  # Correction based on pitch
    return I_base * pitch_correction

# === Hovering Parameters ===
n_hover_rotors = n_forward_rotors
T_per_motor = W / n_hover_rotors  # N
I_hover_motor = hover_current_momentum(T_per_motor, D_m, battery_voltage, rho,
                                       pitch_in, p0, alpha_hover,
                                       eta_prop, eta_motor, I_0)
I_total_hover = I_hover_motor * n_hover_rotors + aux_current
P_total_hover = I_total_hover * battery_voltage
endurance_hover_min = usable_energy_Wh / (P_total_hover / 60)  # minutes

# === Output Parameters
print(f"[Battery Parameters]")
print(f"  Battery Capacity: {battery_capacity_Ah:.2f} Ah")
print(f"  Battery Weight Fraction: {W_batt_empty_frac*100:.2f}%")
print(f"  Aux Current: {aux_current:.2f} A")
print(f"[Aircraft Parameters]")
print(f"  MTOW: {W_MTOW:.4f}g")
print(f"  rho: {rho:.4f}kg/m^3")
print(f"  Sref: {Sref:.4f}m^2")
print(f"  WS_ratio (wing loading): {WS_ratio:.2f}N/m^2")
print(f"  Max Tip Speed: {V_tip:.2f}m/s")
print(f"[Hovering Analysis]")
print(f"  T/W: {T_pred_grams[-1]*3/(W*1000/9.81):.2f}")
print(f"  Required thrust per motor : {T_per_motor:.2f} N")
print(f"  Estimated current per motor : {I_hover_motor:.2f} A")
print(f"  Total hover current (with aux): {I_total_hover:.2f} A")
print(f"  Hover endurance estimate : {endurance_hover_min:.2f} minutes\n")


W  = np.array([1.8,2.0,2.2,2.4]) #kg
V = np.array([15,20,25,30,35,40]) #km/h

A = 0.15*1 #m^2
rho = 1.225 #kg/m^3
mew = 1.81e-5 #kg/m/s

V_m = V*1000/3600 #m/s

W_grid, V_m_grid = np.meshgrid(W, V_m)

#equations
CL_matrix = 2*W_grid / (V_m_grid**2*A*rho)


fig = go.Figure(data=[go.Surface(
    z=CL_matrix,
    x=W_grid,
    y=V_m_grid,
    colorscale='viridis',
    colorbar=dict(title='$C_L$ Scale', len=0.5)
)])
fig.update_layout(
    title='CL vs W and V',
    autosize=False,
    width=800,
    height=560,
    scene=dict(
        xaxis_title='Weight (kg)',
        yaxis_title='Velocity (m/s)',
        zaxis_title='CL'
    ),
    margin=dict(l=65, r=50, b=65, t=90)
)
fig.show()

print("V_m = ", W)
print("CL = ", CL_matrix)

"""
# parse vsp_aero_results.csv
with open('/output/vsp_aero_results.csv', mode='r') as csvfile:
    reader = csv.reader(csvfile)
    header = next(reader)  # Skip header
    results = []
    for row in reader:
        results.append({
            "alpha_deg": float(row[0]),
            "CL": float(row[1]),
            "CD": float(row[2]),
            "CY": float(row[3]),
            "Cl": float(row[4]),
            "Cm": float(row[5]),
            "Cn": float(row[6]),
            "CLtot": float(row[21]),
            "CDtot": float(row[22])
        })
"""
```
